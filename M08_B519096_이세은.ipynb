```python
##### Lab-08-02 : data를 DataFrame 형태로 생성하고 통합하기
###

import pandas as pd

df_left = pd.DataFrame({'주문번호' : ['A101','A102','A103','A104'], '주문일자' : ['2019-4-1','2019-4-2','2019-4-3','2019-5-1'], 
                          '주문자' : ['홍길동','홍진호','홍진영','홍길동']})
df_right= pd.DataFrame({'주문번호' : ['A101','A101','A102','A103','A103','A105'], 
                          '품목명' : ['아이패드','애플펜슬','갤럭시노트7','아이폰X','아이폰케이스','아이폰7'], 
                          '수량' : ['2','1','1','1','2','3']})

left = df_left.merge(df_right, how ='left', on = '주문번호')
right = df_left.merge(df_right, how ='right', on = '주문번호')
outer = df_left.merge(df_right, how ='outer', on = '주문번호')
inner = df_left.merge(df_right, how ='inner', on = '주문번호')

# 왼쪽 데이터는 모두 선택하고 지정된 열(주문번호)에 값이 있는 오른쪽 데이터를 선택
left
```

```python
# 오른쪽 데이터는 모두 선택하고 지정된 열(주문번호)에 값이 있는 왼쪽 데이터를 선택
right
```

```python
# 지정된 열(주문번호)을 기준으로 왼쪽과 오른쪽 데이터를 모두 선택
outer
```

```python
# 지정된 열(주문번호)을 기준으로 왼쪽과 오른쪽 데이터 중 공통 항목만 선택
inner
```
