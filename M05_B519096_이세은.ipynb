```python
##### Lab-05-01 : Calculator 예제 확장
###

## 계산기 클래스
class Calculator:    
    def __init__(self, dNum1, dNum2) :
        self.set_data(dNum1, dNum2)
        self.result = 0
        
    def set_data(self, dNum1, dNum2) :
        self.dNum1 = dNum1
        self.dNum2 = dNum2
        
    def add(self) : 
        self.result = self.dNum1 + self.dNum2
        return self.result
    
    def subtract(self) : 
        self.result = self.dNum1 - self.dNum2
        return self.result

    def multiply(self) : 
        self.result = self.dNum1 * self.dNum2
        return self.result
    
    def divide(self) :
        if self.dNum2 == 0.0 : 
            self.result = 0.0
        else:
            self.result = self.dNum1 / self.dNum2
        return self.result
    
    def exponent(self) :
        self.result = self.dNum1 ** self.dNum2
        return self.result
    
## Derived class
class MaxLimitCal(Calculator):
    # 계산 범위 확인
    def can_Cal(self) :
        result = 0
        
        if (self.dNum1 < -999) : 
            print("첫번째 수가 너무 작습니다. -999보다 큰 숫자를 입력해주세요.")
            result = result + 1          
        elif (self.dNum1 > 10000) : 
            print("첫번째 수가 너무 큽니다. 10000보다 작은 숫자를 입력해주세요.")
            result = result + 1
    
        if (self.dNum2 < -999) : 
            print("두번째 수가 너무 작습니다. -999보다 큰 숫자를 입력해주세요.")
            result = result + 1        
        elif (self.dNum2 > 10000) : 
            print("두번째 수가 너무 큽니다. 10000보다 작은 숫자를 입력해주세요.")
            result = result + 1

        return result
            
# 범위에 맞는 두 실수 입력    
while True:
    dN1 = float(input("첫번째 실수를 입력해주세요: "))
    dN2 = float(input("두번째 실수를 입력해주세요: "))
    print()
    x = MaxLimitCal(dN1, dN2)
    if x.can_Cal() == 0: break;
    print()

# 계산
print("{} + {} = {}".format(x.dNum1, x.dNum2, x.add()))
print("{} - {} = {}".format(x.dNum1, x.dNum2, x.subtract()))
print("{} * {} = {}".format(x.dNum1, x.dNum2, x.multiply()))
print("{} / {} = {}".format(x.dNum1, x.dNum2, x.divide()))
print("{} ^ {} = {}".format(x.dNum1, x.dNum2, x.exponent()))
```

```python
##### Lab-05-02 : Guessing Game 확장
### 컴퓨터가 임의로 생성한 수를 사용자가 맞추는 게임

import random

class game:
    # class variable 생성과 초기화
    finish = 0
    
    def __init__(self, r, N, count):
        self.r = r
        self.N = N
        self.count = count
    
    def guess(self):
        if self.N < self.r:
            print("Guess Bigger!")
        elif self.N > self.r:
            print("Guess Smaller!")
        elif self.N == self.r:
            print("You got it!", end = ' ')
            self.price()
            self.finish = 1
        else:
            print("Error")
            
    # 입력 회수에 따라 차등 출력    
    def price(self):
        if self.count <= 5:
            print("Excellent")
        elif self.count <= 10:
            print("Good")
        else:
            print("Poor, " ,self.count)

## 랜덤 숫자 설정
r = random.randint(0,100)
count = 0
maxCount = 20

# 랜덤 숫자와 비교
while True:
    N = int(input("Guess! : "))
    count += 1
    
    x = game(r, N, count)
    x.guess()
    
    if x.finish == 1: break;    
    if count == maxCount: break;

```
