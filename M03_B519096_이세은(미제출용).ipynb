```python
##### Lab-03-1
### 이차방정식의 근을 구하라!

# 두 근을 읽기
print("aX^2 + bX + c = 0 의 두 근을 구합니다.")
a = float(input("a의 값을 입력해주세요 : "))
b = float(input("b의 값을 입력해주세요 : "))
c = float(input("c의 값을 입력해주세요 : "))

## 판별식
sw = b*b - 4*a*c

# 두 실근을 가지는 경우 
if sw > 0:
    x1 = (-b + sw**0.5)/2
    x2 = (-b - sw**0.5)/2
    print("주어진 이차방정식의 두 실근 : ", x1, ", ", x2)
    
# 중근을 가지는 경우 
if sw == 0:
    x = (-b)/(2*a)
    print("주어진 이차방정식의 중근 : ", x)
    
# 두 허근을 가지는 경우
if sw < 0:
    sw = -sw
    print("주어진 이차방정식의 두 허근 : ",(-b)/2,"+",(sw**0.5)/2,"i, ",(-b)/2,"-",(sw**0.5)/2,"i")


## ex. a=1,b=5,c=6 / a=1,b=-2,c=1 / a=1,b=3,c=3
```

```python
##### Lab-03-2
### 어떤 실수 x에 대한 e^x을 구하라!
# 1 > 18 ~ 171
# 2 > 23
# 3 > 27

# 실수 입력
x = float(input("x의 값을 입력해주세요 : "))

# 변수 초기화
ex = 1
i = 1
fac = 1

# e^x 계산
while i < 100: 
    fac *= i
    ex += ((x**i) / fac)
    i += 1

print(ex)
```

```python
##### Lab-03-3
### 정수 N을 입력 받아 홀수면 3N + 1로, 짝수면 N / 2로 변환하는 과정을 반복하여 1이 되면 프로그램을 종료한다.
# () [] 괄호 모양을 번갈아 출력하기 위해 sw 변수 추가

# 정수 입력
N = int(input("정수값을 입력해주세요 : "))


## 처음 N의 값 출력, 괄호 모양 결정
if N % 2 == 0:                    
    sw = 1
    print("(",N,")", end = ' ')
else:
    sw = 0
    print("[",N,"]", end = ' ')

## N의 값을 변환하고 출력
while N != 1:
    if N % 2 == 1:                  
        N = int(3*N + 1)
    else:                            
        N = int(N/2)   
    if sw == 0:
        print("(",N,")", end = ' ')
        sw = 1
    else:
        print("[",N,"]", end = ' ')
        sw = 0



```

```python
##### Lab-03-4 : Guessing Game
### 컴퓨터가 임의로 생성한 수를 사용자가 맞추는 게임

import random

## 랜덤 숫자 설정
r = random.randint(0,100)
count = 0
maxCount = 6

# 랜덤 숫자와 비교
while True:
    N = int(input("Guess! : "))
    count += 1
    if N < r:
        print("Guess Bigger!")
    elif N > r:
        print("Guess Smaller!")
    elif N == r:
        print("You got it!", end = ' ')
        # 입력 회수에 따라 차등 출력
        if count <= 5:
            print("Excellent")
        elif count <= 10:
            print("Good")
        else:
            print("Poor, " ,count)
        break;
    else:
        print("Error")
    if count == maxCount: break;

```

```python
##### Lab-03-5 : 성적 처리

print("점수를 입력하시면 최고점, 최소점, 평균, 표준편차를 알려드립니다! (입력을 마치시려면 'End'를 입력하세요)")
print()

# 변수 초기화
dAve = 0
dDev = 0
dMax = -9999
dMin = 9999
nCount = 0

# 점수 입력
while True:
    dNum = input("점수를 입력해주세요! : ")
    if dNum == "End":
        break;
    else:
        nCount += 1
        dNum = float(dNum)
        dAve += dNum
        dDev += dNum * dNum
        if dNum > dMax : dMax = dNum;
        if dNum < dMin : dMin = dNum;

# 평균과 표준편차 계산
dAve /= nCount
dDev = ((dDev/nCount)-(dAve)**2)**0.5

# 출력
print()
print("최고점 = ", dMax, " 최소점 = ", dMin, " 평균 = ", dAve, " 표준편차 = ", dDev)
        
        
        
        
        
        
```

```python
##### Lab-03-5 : 성적 처리_확장

#  리스트 생성
student = []
math = []
eng = []
kor = []
soc = []
sci = []
ave = []

print("학생의 이름과 수학,영어,국어,사회,과학 점수를 차례대로 입력해주세요! (입력을 마치시려면 'End'를 입력하세요)")


# 학생과 성적 정보 입력
while True:
    name = input("학생의 이름을 입력하세요 : ")
    if name == "End":
        break;
    else:
        dMath, dEng, dKor, dSoc, dSci = map(float, input("학생의 성적 정보를 입력하세요 : ").split())        
        dAve = (dMath + dEng + dKor + dSoc + dSci)/5
        
        student.append(name)
        math.append(dMath)
        eng.append(dEng)
        kor.append(dKor)
        soc.append(dSoc)
        sci.append(dSci)
        ave.append(dAve)


# 각 학생의 평균 점수 출력 
print()        
for i in range(0, len(student)):
    print(student[i], "학생의 평균 점수는 ", ave[i], "점 입니다.")

    
# 과목의 최고점, 최소점, 평균, 표준편차 구하는 함수
def Compute (lst : list):
    dDev = 0
    for i in range(0, len(lst)):
        n = lst[i]      
        dDev += n*n
    
    dAve = sum(lst) / len(lst)
    dDev = ((dDev/len(lst))-(dAve)**2)**0.5

    return "최고점 = " + str(max(lst)) + " 최소점 = " + str(min(lst)) + " 평균 = " + str(dAve) + " 표준편차 = " + str(dDev)


# 각 과목 최고점, 최소점, 평균 표준편차 출력
print()
print("[수학] ", Compute(math))
print("[영어] ", Compute(eng))
print("[국어] ", Compute(kor))
print("[사회] ", Compute(soc))
print("[과학] ", Compute(sci))
```

```python
##### Lab-03-5 : 성적 처리_확장_객체1

#  리스트 생성
studentList = []
math = []
eng = []
kor = []
soc = []
sci = []

# 학생 클래스
class student:
    def __init__(self, name, math, eng, kor, soc, sci):
        self.name = name
        self.math = math
        self.eng = eng
        self.kor = kor
        self.soc = soc
        self.sci = sci
    def ave(self):
        result = (self.math + self.eng + self.kor + self.soc + self.sci) / 5
        return result

# 과목의 점수를 계산하는 함수        
def Compute (lst : list):
    dDev = 0
    for i in range(0, len(lst)):
        n = lst[i]      
        dDev += n*n
    
    dAve = sum(lst) / len(lst)
    dDev = ((dDev/len(lst))-(dAve)**2)**0.5
    
    return "최고점 = " + str(max(lst)) + " 최소점 = " + str(min(lst)) + " 평균 = " + str(dAve) + " 표준편차 = " + str(dDev)


# 학생의 이름과 성적 정보 입력        
print("학생의 이름과 수학,영어,국어,사회,과학 점수를 차례대로 입력해주세요! (입력을 마치시려면 'End'를 입력하세요)")

while True:
    name = input("학생의 이름을 입력하세요 : ")
    if name == "End":
        break;
    else:
        dMath, dEng, dKor, dSoc, dSci = map(float, input("학생의 성적 정보를 입력하세요 : ").split())        
        # 학생 객체 생성하고 리스트에 추가
        std = student(name, dMath, dEng, dKor, dSoc, dSci)
        studentList.append(std)
        
        math.append(studentList[std].math)
        eng.append(studentList[std].eng)
        kor.append(studentList[std].kor)
        soc.append(studentList[std].soc)
        sci.append(studentList[std].sci)

        
# 학생의 평균점수 출력
print()        
for i in range(0, len(studentList)):
    print( studentList[i].name, "학생의 평균 점수는 ", studentList[i].ave(), "점 입니다.")

# 과목의 계산값 출력
print()
print("[수학] ", Compute(math))
print("[영어] ", Compute(eng))
print("[국어] ", Compute(kor))
print("[사회] ", Compute(soc))
print("[과학] ", Compute(sci))
```

```python
##### Lab-03-5 : 성적 처리_확장_객체2


# 학생 클래스
class student:
    def __init__(self, name, math, eng, kor, soc, sci):
        self.name = name
        self.math = math
        self.eng = eng
        self.kor = kor
        self.soc = soc
        self.sci = sci
    def ave(self):
        result = (self.math + self.eng + self.kor + self.soc + self.sci) / 5
        return result

# 수업 클래스
class subject:
    def __init__(self, name):
        self.name = name
        self.pointList = []
    def Max(self):
        return max(self.pointList)
    def Min(self):
        return min(self.pointList)
    def Ave(self):
        return sum(self.pointList) / len(self.pointList)
    def Dev(self):
        n = 0
        for i in range(0, len(self.pointList)):
            n += (self.pointList[i])**2
        return ((n/len(self.pointList))-(self.Ave())**2)**0.5
    def Info(self):
        return "최고점 = " + str(self.Max()) + " 최소점 = " + str(self.Min()) + " 평균 = " + str(self.Ave()) + " 표준편차 = " + str(self.Dev())
    
    
# 수업 객체 생성
math = subject("수학")
eng = subject("영어")
kor = subject("국어")
soc = subject("사회")
sci = subject("과학")

#  리스트 생성
studentList = []
subjectList = [math, eng, kor, soc, sci]


# 학생 이름과 성적 정보 입력
print("학생의 이름과 수학,영어,국어,사회,과학 점수를 차례대로 입력해주세요! (입력을 마치시려면 'End'를 입력하세요)")

while True:
    name = input("학생의 이름을 입력하세요 : ")
    if name == "End":
        break;
    else:
        dMath, dEng, dKor, dSoc, dSci = map(float, input("학생의 성적 정보를 입력하세요 : ").split())        
        # 학생 객체 생성하고 리스트에 추가
        std = student(name, dMath, dEng, dKor, dSoc, dSci)
        studentList.append(std)
        # 수업 객체에 있는 리스트에 추가
        math.pointList.append(std.math)
        eng.pointList.append(std.eng)
        kor.pointList.append(std.kor)
        soc.pointList.append(std.soc)
        sci.pointList.append(std.sci)

        
# 학생의 평균 점수 출력
print()        
for i in range(0, len(studentList)):
    print( studentList[i].name, "학생의 평균 점수는 ", studentList[i].ave(), "점 입니다.")

# 수업의 최고값, 최소값, 평균, 표준편차 출력
print()
for i in range(0, len(subjectList)):
    print( subjectList[i].Info())
```
